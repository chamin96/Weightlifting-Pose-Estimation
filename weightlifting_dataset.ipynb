```
start = [1251, 1862, 2570, 3012, 4175, 4531, 4860, 5705, 6169, 7334, 9258, 9700, 10208, 10648]
end   = [1253, 1865, 2572, 3014, 4177, 4533, 4862, 5708, 6171, 7335, 9261, 9702, 10210, 10650]

action = 'pullsnatch'
dest = '/content/drive/MyDrive/FYP/wldataset/women'
source = '/content/drive/MyDrive/FYP/wldataset/wl-images'
```

```
from pathlib import Path
Path("{}/{}".format(dest, action)).mkdir(parents=True, exist_ok=True)
```

```
for i in range(len(start)):
  img1 = start[i]
  img2 = end[i]

  for j in range (img1, img2+1) :
    shutil.move('{}/{:08d}.jpg'.format(source, j),
              '{}/{}/{:08d}.jpg'.format(dest, action, j))
```
