```
from pandas import read_csv

# define the dataset location
filename = 'score_report.csv'

# load the csv file as a data frame, header as 1st row and missing values as '?'
dataframe = read_csv(filename, header=0, na_values='?')
```

```
# print first 5 rows in the dataset
dataframe.head()
```

```
# summarize dataset info
dataframe.describe()
```

```
# summarize the shape of the dataset
print(dataframe.shape)
```

```
# create histograms

from matplotlib import pyplot

# select a subset of the dataframe with the chosen columns
k_a = dataframe['kAngle']
b_a = dataframe['bAngle']


# knee angle histograms
k_a.hist(bins=50, figsize=(10,8))
pyplot.xlabel('Angle (degrees)')
pyplot.ylabel('No. of Samples (total = 19')
pyplot.title('Knee Angle Distribution')
pyplot.show()
```

```
# bar angle histograms
b_a.hist(bins=50, figsize=(10,8))
pyplot.xlabel('Angle (degrees)')
pyplot.ylabel('No. of Samples (total = 19')
pyplot.title('Bar Angle Distribution')
pyplot.show()
```

```
fig, (ax1, ax2) = pyplot.subplots(2, figsize=(15, 10))
fig.suptitle('Knee Angle & Knee Score')
ax1.plot(dataframe['kAngle'])
ax2.plot(dataframe['kScore'])
```

```
fig, (ax1, ax2) = pyplot.subplots(2, figsize=(15, 10))
fig.suptitle('Bar Angle & Bar Score')
ax1.plot(dataframe['bAngle'])
ax2.plot(dataframe['bScore'])
```
